```python
# Assuming 'user_by_age' is your DataFrame
plt.figure(figsize=(10, 6))

# Melt the DataFrame to combine 'Users' and 'New users' into a single variable
melted_df = user_by_age.melt(id_vars='Age', value_vars=['Users', 'New users'], var_name='User Type', value_name='Count')

plt.figure(figsize=(12, 8))

# Create a bar plot with stacked bars
sns.barplot(data=melted_df, x='Age', y='Count', hue='User Type', palette=['teal', 'salmon'])

plt.xlabel('Age Category')
plt.ylabel('Number of Users')
plt.title('Age wise users and new users')
plt.legend(title='User Type')

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()



# 18 to 24 age group are high so we should target them
```

```python
#Based on the available data, the "unknown" category appears to have the highest number of users (14,303) and new users (13,636). To better understand and classify this category, further investigation and data analysis are needed. Apart from the "unknown" category, there is a notable introduction of users and new users from the 18-20 age group. 
```

```python
# session engaged and event count

# Set the style of seaborn
sns.set_style("whitegrid")

# Melt it for the grouped bar plot
melted_df = user_by_age.melt(id_vars='Age', value_vars=['Engaged sessions', 'Event count'], var_name='Metric', value_name='Count')

# Create the grouped bar plot
plt.figure(figsize=(12, 8))
sns.barplot(data=melted_df, x='Age', y='Count', hue='Metric', palette=['teal', 'salmon'])

plt.xlabel('Age Category')
plt.ylabel('Count')
plt.title('Age group by sessions organized and events happening')
plt.legend(title='Metric')

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Assuming 'user_by_interest' is your DataFrame containing interest preferences
user_by_interest_sort = user_by_interest.sort_values(by=["Users", "New users"], ascending=False)
top10_user_by_interest = user_by_interest_sort.head(10)

plt.figure(figsize=(12, 8))

# Create bar plot for Users
sns.barplot(data=top10_user_by_interest, x='Interests', y='Users', color='teal', label='Users')

# Create bar plot for New users
sns.barplot(data=top10_user_by_interest, x='Interests', y='New users', color='salmon', label='New Users')

plt.xlabel('Interests')
plt.ylabel('Number of Users / New Users')
plt.title('Top 10 Interest preferences by users and new users')
plt.legend()

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python

# top 10 intrest with the highest engaged session

plt.figure(figsize=(12, 8))

# Create line plot for Engaged sessions
sns.lineplot(data=top10_user_by_interest, x='Interests', y='Engaged sessions', marker='o', color='teal', label='Engaged sessions')

plt.xlabel('Interest')
plt.ylabel('Number of Engaged Sessions')
plt.title('Top 10 Interests with Highest Engaged Sessions')
plt.legend()

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
